Воспользуемся sqllite3 для удобства выполнения заданий

````python
# подключаем SQLite
import sqlite3 as sl
````

````python
# Устанавливаем соединение с базой данных
connection = sl.connect('thecode.db')

# Чтение данных из CSV файла
df = pd.read_csv('hit_albums.csv')

# Запись данных в базу данных SQLite
df.to_sql('hit_albums', connection, if_exists='replace', index=False)

````

````python
#Эта функция принимает объект connection и SELECT-запрос, а возвращает выбранную запись.
# Источник: proglib "Как подружить Python и базы данных"
def execute_read_query(connection, query):
    cursor = connection.cursor()
    result = None
    try:
        cursor.execute(query)
        result = cursor.fetchall()
        return result
    except Error as e:
        print(f"The error '{e}' occurred")
````

Пишем запрос к таблице для вывода только тех артистов у которых как минимум 2 альбома

````python
select_artist = """
SELECT
  *,
  COUNT(Artist) as Альбомы
FROM
  hit_albums
GROUP BY
  Artist
HAVING
  COUNT(Artist) >= 2
ORDER BY
  Artist
"""

artist_album = execute_read_query(connection, select_artist)

for artist_al in artist_album :
  print(artist_al)
# результат запроса загоняем в новый датафрейм
result_aa=pd.read_sql_query(select_artist, connection)
````

Добавим эту таблицу к главной,чтобы можно было увидеть всю информацию по альбому и в последнем столбце количество альбомов у данного артиста (при условии,что альбомов как минимум 2)

````python
select_artist = """
SELECT *
FROM hit_albums h
INNER JOIN (SELECT
                Artist,
                COUNT(Artist)  as Альбомы
            FROM
                hit_albums
            GROUP BY
                Artist
            HAVING
                COUNT(Artist) >= 2
            ORDER BY
                Artist) AS q
          ON h.Artist = q.Artist
ORDER BY h.Artist
"""

artist_album = execute_read_query(connection, select_artist)

for artist_al in artist_album :
  print(artist_al)
# результат запроса загоняем в новый датафрейм
result_aa2=pd.read_sql_query(select_artist, connection)
````

* Создайте новый фрейм данных, содержащий общую сумму всех сертифицированных продаж за каждый год.

С помощью запросы SQL выводим общую сумму сертифицированных продаж за каждый год с сортировкой по годам

````python
select_god = """
SELECT
    Released,
    SUM("Certified sales (millions)") as total_sum
FROM
    hit_albums
GROUP BY Released
ORDER BY Released
"""

god = execute_read_query(connection, select_god)

for god_sum in god :
  print(god_sum)
# результат запроса загоняем в новый датафрейм
result_god=pd.read_sql_query(select_god, connection)
````
